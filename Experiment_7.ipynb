```python
# write a program to find factorial of a number
n = int(input("Enter a number: "))
fact = 1
if n < 0:
  print("Factorial does not exist for negative numbers")
elif n == 0:
  print("The factorial is 1")
elif n > 0:
  while n > 0:
    fact *= n
    n -= 1
  print("Factorial=",fact)
```

```python
# write a program to fibonacci series

n = int(input("Enter a number: "))
a = 0
b = 1
i = 1
while i <= n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i = i + 1
```

```python
# limits in fibonacci series

limit = int(input("Enter a number: "))
a = 0
b = 1
while a <= limit:
  print(a, end = " ")
  a, b = b, a + b
```

```python
# reverse the number
n = int(input("Enter a number: "))
rev = 0
while n > 0:
  digit = n % 10  # Corrected 'num' to 'n'
  rev = rev * 10 + digit
  n = n // 10
print("Reverse=",rev)
```

```python
# check palindrome number/string
# number
num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num = num // 10
if temp == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# string
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True

while i < j:
    if s[i] != s[j]:
        is_palindrome = False
        break
    i += 1
    j -= 1

if is_palindrome:
    print("Yes")
else:
    print("No")
```

```python
# check palindrome string
st = input("Enter the string: ")
rev = st[::-1]
if st == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# count the digits in number
num = int(input("Enter a number: "))
count = 0
while num > 0:
  count += 1
  num = num // 10
print("Number of digits: ", count)
```

```python
# search element in the list
nums = [10,20,30,40,50]
key = int(input("Enter the element to search: "))

i = 0
while i < len(nums):
    if nums[i] == key:
        print("Element found at index", i)
        break
    i += 1
else:
    print("Element not found")
```
